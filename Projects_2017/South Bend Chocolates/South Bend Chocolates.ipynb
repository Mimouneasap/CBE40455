###### 1) Where Does Chocolate Come From? 
Cacao beans, the basis of chocolate, grow in the region 20 degrees north and south of the equator, primarily in South America & Africa. The process of harvesting cacao beans involves hitting the cacao trees with sticks until all the beans fall from the tree. A cacao bean is ripe when it has turned yellow in color and is unripe when it is green, very similar to a banana. 

![](Chocolate/pod.jpg)

Each Cacao bean, contains between 40-60 seeds which are enveloped in a white layer (this becomes cocoa butter). 

![](Chocolate/butter.jpg)

The seeds are removed from this pulp and laid out to dry in the sun. An image of a single bean can be seen below. 

![](Chocolate/bean.jpg)

Once dry, they are buried in a hole and covered with banana leaves. This fermentation process is very important to the development of chocolate because otherwise the chocolate would be bitter and not have the sweet taste we have come to love. The beans are then sent to one of 15 chocolate manufacturers in the U.S. Here they are made into 10 lb bars of milk, dark, and white (cocoa butter) chocolate. These manufacturers supply the chocolate to confectionaries, like the South Bend Chocolate Factory. Specifically, Blommer Chocolate Company is their supplier of chocolate. 

![](Desktop/Chocolate/blommer.jpg)

Upon receiving these chocolate bars, The South Bend Chocolate Factory gets to work. On average, they produce 7,000 lbs of chocolate per day. However, during the Holiday season, they produce as much as 14,000 lbs of chocolate per day. Chocolate is valued based on the actual percentage of cocoa in the pieces, as opposed to added sugar, milk, or other ingredients. Most American chocolate is ~25% cocoa, while Eurpean chocolate is typically 80-90% cocoa. The South Bend Chocolate Factory operates at one of the highest cocoa percentages in America at 75% cocoa. The higher the percentage of cocoa, the better the chocolate is said to be.

###### 3) The Enrobing Room: 
Following the Table Room, we were taken to the Enrobing Room. In this room, there were two enrobers. An enrober is a machine used to cover food items with chocolate. 

![](Chocolate/enrober.jpg)

After being covered with chocolate, the items go through a cooler which blows off extra chocolate and cools the pieces at the same time. In order for the chocolate to be poured over the items it must be melted in the tub above the enrober. If the chocolate is not properly tempered, it could cause the sugars to seep out of the chocolate, causing the pieces to look white and flaky. The chocolates made by this process are pretzels, truffles, and other miscellaneous items. On average, it takes the items approximately two hours to go through the conveyor belt. The South Bend Chocolate Factory has two enrobers. One is used for milk and dark chocolates and is 50 years old, the other is used for white, sugar-free, and seasonal chocolates and is 20 years old.

```python
STN = SBCF

STATES = STN['STATES']
ST_ARCS = STN['ST_ARCS']
TS_ARCS = STN['TS_ARCS']
UNIT_TASKS = STN['UNIT_TASKS']
TIME = STN['TIME']
H = max(TIME)
```

```python
TASKS = set([i for (j,i) in UNIT_TASKS])                         # set of all tasks 

S = {i: set() for i in TASKS}                                    # S[i] input set of states which feed task i
for (s,i) in ST_ARCS:
    S[i].add(s)

S_ = {i: set() for i in TASKS}                                   # S_[i] output set of states fed by task i
for (i,s) in TS_ARCS:
    S_[i].add(s)

rho = {(i,s): ST_ARCS[(s,i)]['rho'] for (s,i) in ST_ARCS}        # rho[(i,s)] input fraction of task i from state s

rho_ = {(i,s): TS_ARCS[(i,s)]['rho'] for (i,s) in TS_ARCS}       # rho_[(i,s)] output fraction of task i to state s

P = {(i,s): TS_ARCS[(i,s)]['dur'] for (i,s) in TS_ARCS}          # P[(i,s)] time for task i output to state s   

p = {i: max([P[(i,s)] for s in S_[i]]) for i in TASKS}           # p[i] completion time for task i

K = {i: set() for i in TASKS}                                    # K[i] set of units capable of task i
for (j,i) in UNIT_TASKS:
    K[i].add(j)
```

## Characterization of States

```python

T = {s: set() for s in STATES}                                   # T[s] set of tasks receiving material from state s 
for (s,i) in ST_ARCS:
    T[s].add(i)

T_ = {s: set() for s in STATES}                                  # set of tasks producing material for state s
for (i,s) in TS_ARCS:
    T_[s].add(i)

C = {s: STATES[s]['capacity'] for s in STATES}   
```

## Characterization of Units

```python

UNITS = set([j for (j,i) in UNIT_TASKS])

I = {j: set() for j in UNITS}                                     # I[j] set of tasks performed with unit j
for (j,i) in UNIT_TASKS:
    I[j].add(i)

Bmax = {(i,j):UNIT_TASKS[(j,i)]['Bmax'] for (j,i) in UNIT_TASKS}  # Bmax[(i,j)] maximum capacity of unit j for task i
Bmin = {(i,j):UNIT_TASKS[(j,i)]['Bmin'] for (j,i) in UNIT_TASKS}  # Bmin[(i,j)] minimum capacity of unit j for task i
```

## Pyomo Model

```python
from pyomo.environ import *
import numpy as np

TIME = np.array(TIME)

model = ConcreteModel()

model.W = Var(TASKS, UNITS, TIME, domain=Boolean)             # W[i,j,t] 1 if task i starts in unit j at time t
model.B = Var(TASKS, UNITS, TIME, domain=NonNegativeReals)    # B[i,j,t,] size of batch assigned to task i in unit j at time t
model.S = Var(STATES.keys(), TIME, domain=NonNegativeReals)   # S[s,t] inventory of state s at time t
model.Q = Var(UNITS, TIME, domain=NonNegativeReals)           # Q[j,t] inventory of unit j at time t
model.Cost = Var(domain=NonNegativeReals)
model.Value = Var(domain=NonNegativeReals)

# Objective is to maximize the value of the final state (see Kondili, Sec. 5)
model.Obj = Objective(expr = model.Value - model.Cost, sense = maximize)

# Constraints
model.cons = ConstraintList()
model.cons.add(model.Value == sum([STATES[s]['price']*model.S[s,H] for s in STATES]))
model.cons.add(model.Cost == sum([UNIT_TASKS[(j,i)]['Cost']*model.W[i,j,t] +
        UNIT_TASKS[(j,i)]['vCost']*model.B[i,j,t] for i in TASKS for j in K[i] for t in TIME])) 

# unit constraints
for j in UNITS:
    rhs = 0
    for t in TIME:
        # a unit can only be allocated to one task 
        lhs = 0
        for i in I[j]:
            for tprime in TIME:
                if tprime >= (t-p[i]+1-UNIT_TASKS[(j,i)]['Tclean']) and tprime <= t:
                    lhs += model.W[i,j,tprime]
        model.cons.add(lhs <= 1)

        # capacity constraints (see Konkili, Sec. 3.1.2)
        for i in I[j]:
            model.cons.add(model.W[i,j,t]*Bmin[i,j] <= model.B[i,j,t])
            model.cons.add(model.B[i,j,t] <= model.W[i,j,t]*Bmax[i,j])

        # unit mass balance
        rhs += sum([model.B[i,j,t] for i in I[j]])
        for i in I[j]:
            for s in S_[i]:
                if t >= P[(i,s)]:
                    rhs -= rho_[(i,s)]*model.B[i,j,max(TIME[TIME <= t-P[(i,s)]])]
        model.cons.add(model.Q[j,t] == rhs)
        rhs = model.Q[j,t]
        
    # terminal condition  
    model.cons.add(model.Q[j,H] == 0)

# state constraints
for s in STATES.keys():
    rhs = STATES[s]['initial']
    for t in TIME:
        # state capacity constraint
        model.cons.add(model.S[s,t] <= C[s])
        
        # state mass balanace
        for i in T_[s]:
            for j in K[i]:
                if t >= P[(i,s)]: 
                    rhs += rho_[(i,s)]*model.B[i,j,max(TIME[TIME <= t-P[(i,s)]])]             
        for i in T[s]:
            rhs -= rho[(i,s)]*sum([model.B[i,j,t] for j in K[i]])
        model.cons.add(model.S[s,t] == rhs)
        rhs = model.S[s,t] 
        
# additional production constraints      
# Forced Variation
    # Pretzel Variation
model.cons.add(model.S['PMPretzel',H]/(model.S['PMPretzel',H]+model.S['PDPretzel',H]+model.S['PWPretzel',H]
                                        +model.S['PSPretzel',H]+model.S['PSFPretzel',H]) >= .1)
model.cons.add(model.S['PDPretzel',H]/(model.S['PMPretzel',H]+model.S['PDPretzel',H]+model.S['PWPretzel',H]
                                        +model.S['PSPretzel',H]+model.S['PSFPretzel',H]) >= .1)
model.cons.add(model.S['PWPretzel',H]/(model.S['PMPretzel',H]+model.S['PDPretzel',H]+model.S['PWPretzel',H]
                                        +model.S['PSPretzel',H]+model.S['PSFPretzel',H]) >= .1)
model.cons.add(model.S['PSPretzel',H]/(model.S['PMPretzel',H]+model.S['PDPretzel',H]+model.S['PWPretzel',H]
                                        +model.S['PSPretzel',H]+model.S['PSFPretzel',H]) >= .1)
model.cons.add(model.S['PSFPretzel',H]/(model.S['PMPretzel',H]+model.S['PDPretzel',H]+model.S['PWPretzel',H]
                                        +model.S['PSPretzel',H]+model.S['PSFPretzel',H]) >= .1)
    # Large Variation
model.cons.add(model.S['PMLarge',H]/(model.S['PMLarge',H]+model.S['PDLarge',H]+model.S['PWLarge',H]
                                        +model.S['PSLarge',H]+model.S['PSFLarge',H]) >= .1)
model.cons.add(model.S['PDLarge',H]/(model.S['PMLarge',H]+model.S['PDLarge',H]+model.S['PWLarge',H]
                                        +model.S['PSLarge',H]+model.S['PSFLarge',H]) >= .1)
model.cons.add(model.S['PWLarge',H]/(model.S['PMLarge',H]+model.S['PDLarge',H]+model.S['PWLarge',H]
                                        +model.S['PSLarge',H]+model.S['PSFLarge',H]) >= .1)
model.cons.add(model.S['PSLarge',H]/(model.S['PMLarge',H]+model.S['PDLarge',H]+model.S['PWLarge',H]
                                        +model.S['PSLarge',H]+model.S['PSFLarge',H]) >= .1)
model.cons.add(model.S['PSFLarge',H]/(model.S['PMLarge',H]+model.S['PDLarge',H]+model.S['PWLarge',H]
                                        +model.S['PSLarge',H]+model.S['PSFLarge',H]) >= .1)
    # Small Variation
model.cons.add(model.S['PMSmall',H]/(model.S['PMSmall',H]+model.S['PDSmall',H]+model.S['PWSmall',H]
                                        +model.S['PSSmall',H]+model.S['PSFSmall',H]) >= .1)
model.cons.add(model.S['PDSmall',H]/(model.S['PMSmall',H]+model.S['PDSmall',H]+model.S['PWSmall',H]
                                        +model.S['PSSmall',H]+model.S['PSFSmall',H]) >= .1)
model.cons.add(model.S['PWSmall',H]/(model.S['PMSmall',H]+model.S['PDSmall',H]+model.S['PWSmall',H]
                                        +model.S['PSSmall',H]+model.S['PSFSmall',H]) >= .1)
model.cons.add(model.S['PSSmall',H]/(model.S['PMSmall',H]+model.S['PDSmall',H]+model.S['PWSmall',H]
                                        +model.S['PSSmall',H]+model.S['PSFSmall',H]) >= .1)
model.cons.add(model.S['PSFSmall',H]/(model.S['PMSmall',H]+model.S['PDSmall',H]+model.S['PWSmall',H]
                                        +model.S['PSSmall',H]+model.S['PSFSmall',H]) >= .1)
    # Popcorn Variation
model.cons.add(model.S['PMPopcorn',H]/(model.S['PMPopcorn',H]+model.S['PDPopcorn',H]+model.S['PWPopcorn',H]
                                        +model.S['PSPopcorn',H]+model.S['PSFPopcorn',H]) >= .1)
model.cons.add(model.S['PDPopcorn',H]/(model.S['PMPopcorn',H]+model.S['PDPopcorn',H]+model.S['PWPopcorn',H]
                                        +model.S['PSPopcorn',H]+model.S['PSFPopcorn',H]) >= .1)
model.cons.add(model.S['PWPopcorn',H]/(model.S['PMPopcorn',H]+model.S['PDPopcorn',H]+model.S['PWPopcorn',H]
                                        +model.S['PSPopcorn',H]+model.S['PSFPopcorn',H]) >= .1)
model.cons.add(model.S['PSPopcorn',H]/(model.S['PMPopcorn',H]+model.S['PDPopcorn',H]+model.S['PWPopcorn',H]
                                        +model.S['PSPopcorn',H]+model.S['PSFPopcorn',H]) >= .1)
model.cons.add(model.S['PSFPopcorn',H]/(model.S['PMPopcorn',H]+model.S['PDPopcorn',H]+model.S['PWPopcorn',H]
                                        +model.S['PSPopcorn',H]+model.S['PSFPopcorn',H]) >= .1)
# Turtle Variation
model.cons.add(model.S['PMTurtles',H]/(model.S['PMTurtles',H]+model.S['PDTurtles',H]) >= .3)


SolverFactory('gurobi').solve(model).write()
```
