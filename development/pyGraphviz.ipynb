```python
import graphviz

dot = graphviz.Digraph()
dot.node('a',shape='circle')
dot.node('b',shape='circle')
dot
```

```python
dot.edge('a','b')
dot.edge('b','a')
dot
```

```python
clp = graphviz.Digraph()
clp.attr(rankdir='LR')
clp.node('+',shape='circle',pos=(1,1))
clp.node('K',shape='rect')
clp.node('G',shape='rect')

clp.edge('+','K')
clp.edge('K','G')
clp.edge('G','+')

clp
```
