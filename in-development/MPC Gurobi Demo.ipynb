```python
from gurobipy import *
import numpy as np


m = Model('MPC Demo')


dt = 0.1
T = 10

tgrid = np.linspace(0,T,(T/dt)+1)
    
N = len(tgrid)


h1 = m.addVars(N,name="H1")
u = m.addVars(N)

m.addConstr(h1[0] == 0)


for k in range(0,N-1):
    m.addConstr(h1[k+1] == h1[k] + dt*(u[k] - 0.2*h1[k])/2.0)
    
    
p = sum([abs(h-1) for h in h1])
m.setObjective(p, GRB.MINIMIZE)

m.update

m.optimize()
h1
```
