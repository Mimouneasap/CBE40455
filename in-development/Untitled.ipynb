```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from scipy.stats import norm

plt.figure(figsize=(10,4))
plt.xlabel('Trading Days')

N = 2000    # number of realizations
T = 252    # simulation time horizon in trading days
h = 1      # time step in trading days

r = 0.04/252.0
sigma = 0.6/np.sqrt(252.0)

# Initialize P
S = np.array([100.0]*N)

# generator of random variates
Zgenerator = (np.array(norm.rvs(loc=r*h,scale=sigma*np.sqrt(h),size=N)) for k in range(0,T))

k = 252
for Z in Zgenerator:
    k += 1
    S += S*Z
    plt.plot([k]*N,S,'b.',ms=1)
```
